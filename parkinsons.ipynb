## Description of the problem

### parkinson's disease

A disorder of the central nervous system that affects movement, often including tremors. <br/>
Nerve cell damage in the brain causes dopamine levels to drop, leading to the symptoms of Parkinson's. <br/>
Parkinson's often starts with a tremor in one hand. Other symptoms are slow movement, stiffness and loss of balance. <br/>
The rarity of the disease is highly common with more than one million cases per year in india.

#### Data description

*name* - ASCII subject name and recording number <br/>
*MDVP:Fo(Hz)* - Average vocal fundamental frequency <br/>
*MDVP:Fhi(Hz)* - Maximum vocal fundamental frequency <br/>
*MDVP:Flo(Hz)* - Minimum vocal fundamental frequency <br/>
*MDVP:Jitter(%)*, <br/>
*MDVP:Jitter(Abs)*, <br/>
*MDVP:RAP*, <br/>
*MDVP:PPQ*, <br/>
*Jitter:DDP* - Several measures of variation in fundamental frequency <br/>
*MDVP:Shimmer*, <br/>
*MDVP:Shimmer(dB)*, <br/>
*Shimmer:APQ3*, <br/>
*Shimmer:APQ5*, <br/>
*MDVP:APQ*, <br/>
*Shimmer:DDA -* Several measures of variation in amplitude <br/>
*NHR*, *HNR -* Two measures of ratio of noise to tonal components in the voice <br/>
*status -* Health status of the subject (one) - Parkinson's, (zero) - healthy <br/>
*RPDE*, D2 -* Two nonlinear dynamical complexity measures <br/>
*DFA -* Signal fractal scaling exponent <br/>
*spread1,spread2,PPE -* Three nonlinear measures of fundamental frequency variation <br/>

## Model Selection,Training and Evaluation

We will be using XGBoost, which stands for “Extreme Gradient Boosting” and it has become one of the most popular </br>
and widely used machine learning algorithms due to its ability to handle large datasets and </br>
its ability to achieve state-of-the-art performance in many machine learning tasks such as </br>
classification and regression

### hyperparameter tuning

we will fine tune the hyperparameters with gridsearchcv 

```python
import xgboost as xgb
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import precision_recall_curve, confusion_matrix
```

```python
X_, Xtest, y_, ytest = train_test_split(X, y, test_size=0.20, random_state=0, stratify=y)
# the test set will be later used to score the fully trained model as to provide a generalized score
```

```python
params = {
          'gamma': [0.1,0.5,3.2,12.8,51.2,102.4, 200],
          'learning_rate': [0.01, 0.06, 0.1, 0.2, 0.4, 0.7],
          'max_depth': [6,8,10,12,14],
          'n_estimators': [50,80,115,150]
          }
```

Since with early detection of parkinson's disease, treatment, and expert care, the patient can live a long, </br>
productive, and meaningful life and with many inexpensive tests out there to diagnose parkinson's disease, </br> 
it is clear that we want to avoid false negative with some tolerance in false postives </br> 
*i.e* recall(sensitivity) must be very close to 1 with some tolerance in precision score.

```python
model_instance = xgb.XGBClassifier(objective='binary:logistic',
                                  booster='gbtree')
clf = GridSearchCV(estimator=model_instance, param_grid=params, cv=3, scoring='average_precision', verbose=1)
clf.fit(X_, y_)
# we are using average_precision_score which approximates the auc-pr(area under the curve-precision recall curve)
# as with higher auc-pr the greater the recall score when the precision score starts to fall off
```

```python
clf.best_params_   # we will be using the following hyperparameters henceforth.
```

with different threshold we may get different values of precision/recall, we want to find the optimum threshold for our </br>
problem. For that we will start by plotting the pr(precision-recall) curve for different values of threshold

```python
Xtrain, Xval, ytrain, yval = train_test_split(X_, y_, test_size=0.25, random_state=26)

xg_model = xgb.XGBClassifier(objective='binary:logistic',
                         booster='gbtree',
                            gamma=0.1,
                            learning_rate=0.1,
                            max_depth=6,
                            n_estimators=80)
xg_model.fit(Xtrain, ytrain)
```

```python
yproba_val = xg_model.predict_proba(Xval)
precision, recall, thresholds = precision_recall_curve(yval, yproba_val[:,1])
thresholds_mod = np.concatenate([thresholds,[1+max(thresholds)]])
```

```python
plt.plot(recall, precision, marker='x', color='red')
for pre, re, threshold in np.nditer([precision, recall, thresholds_mod]):
    value = "{:.2f}".format(threshold)
    plt.annotate(text=value, xy = (re,pre), textcoords='offset points', xytext=(10,50), size=7, 
                 arrowprops=dict(facecolor='green', alpha=0.5, arrowstyle='fancy',connectionstyle='angle3'))
plt.xlabel('recall')
plt.ylabel('precision')
plt.title('Precision-Recall Curve', loc='right');
```

#### <font color = 'green'> The values marked by the arrows give the probability threshold values </font> </br>
with the exception of the first threshold which is close to 2 as to classifying everyone as negative which </br>
gives *precision = 1* and *recall=0*

```python
optimal_threshold = thresholds[np.argmin(abs(thresholds-0.11))]
optimal_threshold
```

```python
ypred_val = (yproba_val[:,1]>=optimal_threshold).astype(int)

con_mat_val = confusion_matrix(yval, ypred_val)
sns.heatmap(con_mat_val, annot=True, cbar=False)
plt.xlabel('Predicted Label')
plt.ylabel('Actual Label')
plt.title('Confusion matrix for validation set');
```

```python
from sklearn.metrics import classification_report, accuracy_score

print(classification_report(yval, ypred_val))
```

### Model generalization(testing)

```python
final_model = xgb.XGBClassifier(objective='binary:logistic',
                         booster='gbtree',
                            gamma=0.1,
                            learning_rate=0.1,
                            max_depth=6,
                            n_estimators=80)
final_model.fit(X_, y_)
```

```python
yproba_test = final_model.predict_proba(Xtest)
yproba_test = yproba_test[:,1]
ypred_test = (yproba_test>=optimal_threshold).astype(int)
```

```python
con_mat_test = confusion_matrix(ytest, ypred_test)
sns.heatmap(con_mat_test, annot=True, cbar=False)
plt.xlabel('Predicted Label')
plt.ylabel('Actual Label')
plt.title('Confusion matrix for testing set');
```

```python
print(classification_report(ytest, ypred_test))
```

 thus we can see our model is working good with *recall score* of 1.00 and *precision score* of 0.94 </br>
 which gives *f1_score* of 0.97 (which is the harmonic mean of recall and precision) along with </br>
 *accuracy score* of 0.95
